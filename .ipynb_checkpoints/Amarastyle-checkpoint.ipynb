```python
joke = "That's what she said"
c1 = 0
c2 = 0
ids = []
for row in scripts_df.itertuples():
    if joke.lower() in row[5].lower():
        c1 += 1
        ids.append(row[0])
    if "she said" in row[5]:
        c2 += 1
    add_character(row[6], characters_data)
    add_season(row[2], row[6], characters_data)
    add_episode(row[3], row[2], row[6], characters_data)

print("insensitive:", c1, "sensitive", c2)
for key in ids:
    print(scripts_df["line_text"].iloc[key])
print(len(characters_data.keys()))
```
